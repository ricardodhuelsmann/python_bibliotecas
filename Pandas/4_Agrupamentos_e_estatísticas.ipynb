#Pandas
###Cálculos e variáveis estatísticas

```python
import pandas as pd
df = pd.read_csv('pokemon_dados_modificados.csv')
```

###Função 'Groupby'

```python
#Classificar os valores de acordo com o Tipo, apresentando a média de cada parâmetro
df.groupby(['Type 1']).mean()
```

```python
#Classificar os valores de acordo com o Tipo 2, apresentando a média de cada parâmetro, em ordem descrecente de ataque
df.groupby(['Type 2']).mean().sort_values('Attack', ascending=False)
```

```python
#Classificar e contar por determinado parâmetro

df['Contador'] = 1
resumo_t1 = df.groupby(['Type 1']).count().sort_values('Contador', ascending=False)['Contador']
print(resumo_t1)
```

```python
#Classificar e agrupar de acordo com grupo dentro de grupo
resumo_t1_t2 = df.groupby(['Type 1', 'Type 2']).count()['Contador']
print(resumo_t1_t2)
```
