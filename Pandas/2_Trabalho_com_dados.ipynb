#Pandas
####Editando e criando dados

```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

###Criando uma nova coluna

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(10)
```

###Removendo uma coluna

```python
df = df.drop(columns=['Total'])
df.head(10)
```

###Criando de outra maneira uma coluna


```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1) #Soma em todas as linhas, as colunas 4 até 9, linha a linha
df.head(10)
```

###Reordenando as colunas

```python
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] #Insere a última coluna na posição desejada
df.head(10)
```

###Salvando a tabela modificada

```python
df.to_csv('pokemon_dados_modificados.csv', index=False)
```
