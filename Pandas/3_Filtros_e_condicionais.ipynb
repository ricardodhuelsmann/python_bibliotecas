#Pandas
###Filtrando dados

```python
import pandas as pd
df = pd.read_csv('pokemon_dados_modificados.csv')
```

###Filtrando dados por tipo 

```python
df.loc[df['Type 1'] == 'Water']
```

```python
df.loc[df['Type 1'] == 'Water'].describe()
```

###Filtrando por duas ou mais classes, e salvando em novo DataFrame


```python
novo_df = df.loc[(df['Type 1'] == 'Water') & (df['HP'] > 100) & (df['Attack'] < 100)]
novo_df
```

```python
novo_df.describe()
```

###Resetar índice do DataFrame

```python
novo_df = novo_df.reset_index(drop=True)
novo_df
```

###Filtros avançados (Regular Expressions - re)

```python
import re
```

```python
df_filtrado = df.loc[df['Name'].str.contains('Mega')]
df_filtrado
```

```python
df_filtrado = df.loc[df['Type 1'].str.contains('water|fire', regex=True)]
df_filtrado
```

```python
#Ignora upper case
df_filtrado = df.loc[df['Type 1'].str.contains('water|fire', flags= re.I, regex=True)]
df_filtrado
```

```python
#Encontrar nomes que comecem com "Pi"
df_filtrado = df.loc[df['Name'].str.contains('^pi[a-z]*', flags= re.I, regex=True)]
df_filtrado
```

###Troca condicional

```python
#Pode ser uma alternativa para traduzir tabelas
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Fogo'
df.loc[df['Type 1'] == 'Water', 'Type 1'] = 'Água'
df.loc[df['Type 1'] == 'Grass', 'Type 1'] = 'Grama'
df.loc[df['Type 1'] == 'Rock', 'Type 1'] = 'Pedra'
df
```

```python
#Cria um atributo, e atribui valor dependendo de condicional
df.loc[df['HP'] > 50, ['Novo atributo']] = 'TOP'
df.loc[df['HP'] <= 50, ['Novo atributo']] = 'low'
df
```

```python
#Removendo coluna adicionada
df = df.drop(columns=['Novo atributo'])
df
```

```python
#Adicionando valores a duas colunas simultaneamente
df.loc[df['Total'] > 500, ['Atributo 1', 'Atributo 2']] = ['TOP', 'EXTRA']
df
```
