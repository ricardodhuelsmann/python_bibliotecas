#Matplotlib
###Criando e editando gráficos de dispersão

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
'''
Criando dados de dispersão

'''
valores_x = np.arange(100)
valores_y = np.random.randn(100)*10+valores_x+10

plt.plot(valores_x, valores_y, linestyle='none', marker='o')
plt.show()
```

```python
'''
Editando o gráfico
'''
plt.plot(valores_x, valores_y, linestyle='none', marker='H', 
         markerfacecolor='#8080ff', markeredgecolor="#00004d", markeredgewidth=2,  markersize=10)
plt.figure(figsize=(8,6), dpi=300)

plt.show()
```

```python
'''
Adicionando marcadores
'''

plt.plot(valores_x, valores_y, linestyle='none', marker='H', 
         markerfacecolor='#8080ff', markeredgecolor="#00004d", markeredgewidth=2,  markersize=10)

plt.annotate('Observe aqui!', xy=(25, 50), xytext=(0, 80), arrowprops=dict(facecolor='#8080ff', shrink=0.05))

plt.show()
```

```python
'''
Adicionando linhas
'''

plt.plot(valores_x, valores_y, linestyle='none', marker='H', 
         markerfacecolor='#8080ff', markeredgecolor="#00004d", markeredgewidth=2,  markersize=10)

plt.axvline(54, color='k')
plt.axhline(85, color='b')

plt.show()
```

```python
import matplotlib.patches as patches
'''
Adicionando formas de foco
'''
figura = plt.figure()
eixo = figura.add_subplot(111)
eixo.plot(valores_x, valores_y, linestyle='none', marker='H',
          markerfacecolor='#8080ff', markeredgecolor="#00004d", markeredgewidth=2,  markersize=10)
 
# Retângulo
eixo.add_patch(patches.Rectangle((0, 0), 40, 40, alpha=0.5, facecolor="cyan", edgecolor="black", linewidth=3, linestyle='solid'))
#Círculo
eixo.add_patch(patches.Circle((50, 50), 20, alpha=0.2, facecolor="b", edgecolor="black", linewidth=3, linestyle='solid'))
#Elipse
eixo.add_patch(patches.Ellipse((90, 100), 20, 120, 10, alpha=0.4, facecolor="r", edgecolor="black", linewidth=3, linestyle='solid'))



# Show the graph
plt.show()
```

```python
'''
Plotando e linearizando dispersão
'''

plt.plot(valores_x, valores_y, linestyle='none', marker='H', 
         markerfacecolor='#8080ff', markeredgecolor="#00004d", markeredgewidth=2,  markersize=10)

b, a = np.polyfit(valores_x, valores_y, deg=1)

xseq = np.linspace(0, 100, num=101)

plt.plot(xseq, a + b * xseq, color="k", lw=2.5)

a1 = float("{:.2f}".format(a))
b1 = float("{:.2f}".format(b))
r = np.corrcoef(valores_x, valores_y)
r = r[0,1]*r[0,1]
r = float("{:.4f}".format(r))

plt.text(50, 12, f'f(x): ${a1}x + {b1}$', fontsize=15)
plt.text(62, 0, f'r²: ${r}$', fontsize=12)

plt.show()
```
