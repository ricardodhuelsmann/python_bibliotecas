#Matplotlib
###Introdução ao Matplotlib

```python
'''
Inicialmente é necessario importar a biblioteca
Matplotlib, assim como algumas outras necessárias
para trabalhar com dados (NumPy e Pandas)
'''
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
'''
Criação de um gráfico básico
'''

x = [1,2,3]
y = [2,4,6]

plt.plot(x, y)
plt.show()
```

```python
'''
Adicionando informações básicas ao gráfico
'''

plt.plot(x, y)
plt.title('Título do gráfico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.show()
```

```python
'''
Editando formas das informações no gráfico
'''
plt.plot(x, y)
plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.show()
```

```python
'''
Editando a escala
'''
plt.plot(x, y)
plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])

plt.show()
```

```python
'''
Adicionando legenda
'''
plt.plot(x, y, label='Dados iniciais') #Adicionar label
plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])

plt.legend(fontsize=14, loc=2) #Localização 2 significa superior esquerda

plt.show()
```

```python
'''
Alterar forma da linha de dados
'''
plt.plot(x, y, label='Dados iniciais', color='red', linestyle='--',
         linewidth=4, marker='o', markersize=12, markeredgecolor='black') #Adicionar informações

plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])
plt.legend(fontsize=14, loc=2) 
plt.show()
```

```python
'''
Existe uma maneira de abreviar a formatação dos dados do gráfico
O seguinte modo é utilizado
[cor][marcador][linha]
'''

plt.plot(x, y, 'ro--', label='Dados iniciais', linewidth=4, 
         markersize=12, markeredgecolor='black') #Adicionar informações abreviadas

plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])
plt.legend(fontsize=14, loc=2) 
plt.show()
```

```python
'''
Adicionando mais dados ao mesmo gráfico
'''
x2 = np.arange(1.25, 3, 0.25) #Cria um array de 0.25 até 2.75, com acréscimo de 0.25
plt.plot(x2, x2**2-6*x2+11, 'k^-', label='Novos dados',
         linewidth=4,markersize=12, markeredgecolor='blue') #É possível passar Y como função de X

plt.plot(x, y, 'ro--', label='Dados iniciais', linewidth=4, 
         markersize=12, markeredgecolor='black') #Adicionar informações abreviadas
plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])
plt.legend(fontsize=14, loc=2) 
plt.show()
```

```python
'''
Alterando o tamanho da figura gerada
'''
plt.figure(figsize=(8,5), dpi=100) #dpi são'dots per inch', 300 dpi corresponde a alta qualidade

plt.plot(x2, x2**2-6*x2+11, 'k^-', label='Novos dados',
         linewidth=4,markersize=12, markeredgecolor='blue') 
plt.plot(x, y, 'ro--', label='Dados iniciais', linewidth=4, 
         markersize=12, markeredgecolor='black')
plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])
plt.legend(fontsize=14, loc=2) 
plt.show()
```

```python
'''
Salvando a figura gerada
'''

plt.figure(figsize=(8,5), dpi=100) 
plt.plot(x2, x2**2-6*x2+11, 'k^-', label='Novos dados',
         linewidth=4,markersize=12, markeredgecolor='blue') 
plt.plot(x, y, 'ro--', label='Dados iniciais', linewidth=4, 
         markersize=12, markeredgecolor='black')
plt.title('Título do gráfico', fontdict={'fontname': 'Helvetica', 'fontsize': 20})
plt.xlabel('Eixo X', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.ylabel('Eixo Y', fontdict={'fontname': 'Helvetica', 'fontsize': 16})
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 5, 6, 7])
plt.legend(fontsize=14, loc=2)

plt.savefig('meu_grafico.png', dpi=300) #Salvando em alta resolução

plt.show()
```
