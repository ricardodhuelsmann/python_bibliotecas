#Matplotlib
###Criando histogramas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
d1 = np.random.randint(0, high=100, size=1000)
d2 = np.random.randint(0, high=100, size=1000)
dados = d1+d2

plt.hist(dados, bins=10, edgecolor="black")
plt.show()
```

```python
'''
Definindo região e editando
'''
eixo_limitado = np.arange(50, 175, 20)
mediana = np.median(dados)

plt.hist(dados, bins=eixo_limitado, edgecolor="white", alpha=0.5)
plt.axvline(x=mediana, color="black", ls="--", label="Mediana")

plt.xlabel('Dados')
plt.ylabel('Frequência')
plt.legend()

plt.show()
```

###Gráficos pizza

```python
valores_pizza = [10, 15, 20, 25]

plt.pie(valores_pizza)
plt.show()
```

```python
'''
Adicionando rótulos
'''

nomes = ['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4']

plt.pie(valores_pizza, labels = nomes, labeldistance=1.15)
plt.show()
```

```python
'''
Editando
'''
cores = ['#9999ff', '#6666ff', '#3333ff', '#0000cc']

plt.pie(valores_pizza, labels = nomes, labeldistance=1.15, wedgeprops = { 'linewidth' : 2, 'edgecolor' : 'white' }, 
        colors = cores, startangle = 180, counterclock = False, shadow = True, textprops={'fontsize': 14})

plt.title('Gráfico pizza', fontdict={'fontsize': 20})

plt.show()
```
